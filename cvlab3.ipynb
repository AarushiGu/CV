```python
img1=cv2.imread('/content/coarse.jpg') #coarse
cv2_imshow(img1)
print()
gray_image1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image1)
```

```python
img2=cv2.imread('/content/random.jpg') #random
cv2_imshow(img2)
print()
gray_image2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image2)
```

```python
from skimage.feature import greycomatrix, greycoprops
```

```python
# Apply Gaussian filter
coarse_img_filtered = cv2.GaussianBlur(gray_image1, (5, 5), 0)
smooth_img_filtered = cv2.GaussianBlur(gray_image, (5, 5), 0)
random_img_filtered = cv2.GaussianBlur(gray_image2, (5, 5), 0)
```

```python
# Calculate GLCM (Gray Level Co-occurrence Matrix) for each image
distances = [1]
angles = [0]
properties = ['contrast', 'energy', 'correlation', 'homogeneity']

def calculate_texture_features(image):
    glcm = greycomatrix(image, distances=distances, angles=angles, levels=256, symmetric=True, normed=True)
    texture_features = [greycoprops(glcm, prop).ravel()[0] for prop in properties]
    return texture_features


coarse_features = calculate_texture_features(coarse_img_filtered)
smooth_features = calculate_texture_features(smooth_img_filtered)
random_features = calculate_texture_features(random_img_filtered)
```

```python
print("Coarse texture features:")
print("Contrast:", coarse_features[0])
print("Energy:", coarse_features[1])
print("Correlation:", coarse_features[2])
print("Homogeneity:", coarse_features[3])


print("\nSmooth texture features:")
print("Contrast:", smooth_features[0])
print("Energy:", smooth_features[1])
print("Correlation:", smooth_features[2])
print("Homogeneity:", smooth_features[3])


print("\nRandom texture features:")
print("Contrast:", random_features[0])
print("Energy:", random_features[1])
print("Correlation:", random_features[2])
print("Homogeneity:", random_features[3])
```
