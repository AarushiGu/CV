```python
gray_image=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#scaling
img_shrinked = cv2.resize(gray_image, (200, 250),
                         interpolation=cv2.INTER_AREA) #inter_cubic for enlarging the image
cv2_imshow(img_shrinked)
```

```python
img_shrinked = cv2.resize(gray_image, (400, 200),
                         interpolation=cv2.INTER_AREA) #inter_cubic for enlarging the image
cv2_imshow(img_shrinked)
```

```python
#translation
rows,cols=gray_image.shape #to not crop the image from both sides
M = np.float32([[1, 0, 100], [0, 1, 50]]) #x translation to right,left,units y translation to up,down,units
dst = cv2.warpAffine(gray_image,M,(cols,rows))
cv2_imshow(dst)
```

```python
#rotation
image = cv2.rotate(gray_image, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(image)
```

```python
#shearing in x-axis
# the boundaries of the image that are parallel to the x-axis keep their location,
#and the edges parallel to the y-axis change their place depending on the shearing factor.
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(gray_image, M,
                                 (int(cols*1.5),
                                 int(rows*1.5)))
cv2_imshow(sheared_img)
```

```python
# Create a black background
image = np.zeros((400, 400), dtype=np.uint8)

# Define the vertices of the right-angle triangle
pts = np.array([[0, 0], [0, 400], [400, 0]])

# Draw the white triangle on the black background
cv2.fillPoly(image, [pts], 255)

# Display the image
cv2_imshow(image)
```

```python
# Define dimensions of the chessboard
rows = 8
cols = 8
square_size = 50

# Create a blank white canvas
canvas = np.ones((rows * square_size, cols * square_size), dtype=np.uint8) * 255

# Fill the canvas with grey and black squares
for i in range(rows):
    for j in range(cols):
        if (i + j) % 2 == 0:
            color = 100  # Grey
        else:
            color = 0    # Black
        x1, y1 = j * square_size, i * square_size
        x2, y2 = x1 + square_size, y1 + square_size
        cv2.rectangle(canvas, (x1, y1), (x2, y2), color, -1)

# Display the chessboard pattern
cv2_imshow( canvas)
```

```python
# Define the dimensions of the image
height, width = 400, 400

# Create a black background image
image = np.zeros((height, width), dtype=np.uint8)

# Define the center and radius of the circle
center = (width // 2, height // 2)
radius = min(width, height) // 4

# Draw a white circle on the black background
cv2.circle(image, center, radius, (255), thickness=-1)

# Display the image
cv2_imshow(image)
```

```python
# Define the dimensions of the image
height = 400
width = 600

# Create a black background
image = np.zeros((height, width), dtype=np.uint8)

# Define the rectangle parameters
rectangle_color = (255)  # White color
rectangle_thickness = -1  # Fill the rectangle completely
rectangle_top_left = (100, 100)  # Top-left corner coordinates
rectangle_bottom_right = (500, 300)  # Bottom-right corner coordinates

# Draw the white rectangle on the black background
cv2.rectangle(image, rectangle_top_left, rectangle_bottom_right, rectangle_color, rectangle_thickness)

# Display the image
cv2_imshow(image)
```
