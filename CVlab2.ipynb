```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/abc.jpg")
plt.imshow(img)
```

```python
# Creating kernel
kernel = np.ones((5, 5), np.uint8)
_,mask=cv2.threshold(img,0,255,cv2.THRESH_BINARY_INV)
# Using cv2.erode() method
image_erode = cv2.erode(mask, kernel,iterations=1)
plt.imshow(image_erode)
```

```python
image_dilation = cv2.dilate(mask, kernel, iterations=1)
plt.imshow(image_dilation)
```

```python
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel) #erosion followed by dilation
plt.imshow(opening)
```

```python
closing=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel) #Dilation followed by Erosion
plt.imshow(closing)
```

```python
#task2
img2=cv2.imread("/content/fingerprint.png")
plt.imshow(img2)
```

```python
kernel = np.ones((5,5), np.uint8)
_,mask=cv2.threshold(img2,8,255,cv2.THRESH_BINARY) #masking
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel) #erosion followed by dilation
plt.imshow(opening)   # It is generally used to remove the noise in the image
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img2, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(opening, cmap='gray')
plt.title('opening')
```

```python
#task3
img3=cv2.imread("/content/task3.png")
plt.imshow(img3)
```

```python
# Define the kernel for dilation and erosion
kernel = np.ones((3, 3), np.uint8)

# Grayscale Dilation
dilation = cv2.dilate(img3, kernel, iterations=1)

# Grayscale Erosion
erosion = cv2.erode(img3, kernel, iterations=1)

# Morphological Gradient
gradient = cv2.absdiff(dilation, erosion)

# Display the original, dilation, erosion, and gradient images
plt.subplot(1, 4, 1)
plt.imshow(img3, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')

plt.subplot(1, 4, 3)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')

plt.subplot(1, 4, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')

# Adjust layout for better display
plt.tight_layout()

# Show the plot
plt.show()
```
